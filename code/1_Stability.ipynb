# Plant location model

In the classical plant location model, we have a network, consisting of a set $\mathcal{P}$ of potential source nodes and a set $\mathcal{D}$ of demand nodes:

- for each $i \in \mathcal{P}$ we have a fixed opening cost 
$f_i$ and a capacity level $R_i$;
- for each $j \in \mathcal{D}$ we have a demand $d_j$ (we assume a single product type for simplicity);
- for each arc $(i,j)$ in the network we have a unit transportation cost $c_{ij}$.

The aim is satisfying demand at minimum cost.

$$
\begin{array}
\text{min} \quad & \sum_{i \in \mathcal{P}} f_i y_i + \sum_{i \in \mathcal{P}} \sum_{j \in \mathcal{D}} c_{ij} x_{ij} \\
\text{st} \quad & \sum_{i \in \mathcal{P}} x_{ij} \geq d_j, && \forall j \in \mathcal{D} \\
& \sum_{j \in \mathcal{D}} x_{ij} \leq R_i y_i, && \forall i \in \mathcal{P}   \\
& x_{ij} \geq 0, \; y_i \in \{ 0,1 \} 
\end{array}
$$

```python
# import library
import gurobipy as gp
from gurobipy import GRB
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Constants
cost_factor = 10  # Constant to scale the distance into cost
num_customers = 5
num_plants = 3
np.random.seed(1)  # For reproducibility

# Generate random positions for plants and customers on a 2D plane
customer_positions = {i: (np.random.rand(), np.random.rand()) for i in range(num_customers)}
plant_positions = {j: (np.random.rand(), np.random.rand()) for j in range(num_plants)}

# Function to calculate Euclidean distance between two points
def euclidean_distance(p1, p2):
    return np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

# Compute transportation costs as a constant times the Euclidean distance
transport_cost = [
    [cost_factor * euclidean_distance(customer_positions[i], plant_positions[j]) for j in range(num_plants)]
    for i in range(num_customers)
]

# Fixed cost for opening each plant (randomized for variety)
fixed_cost = [1000 + 200 * j for j in range(num_plants)]
```

```python
# Create the bipartite graph
B = nx.Graph()

# Add plant nodes
plant_nodes = ['Plant ' + str(j) for j in range(num_plants)]
B.add_nodes_from(plant_nodes, bipartite=0)  # Add plants to set 0 of the bipartite graph

# Add customer nodes
customer_nodes = ['Customer ' + str(i) for i in range(num_customers)]
B.add_nodes_from(customer_nodes, bipartite=1)  # Add customers to set 1 of the bipartite graph

# Prepare node colors: blue for open plants, red for customers
node_colors = []
for node in B.nodes():
    if 'Plant' in node:
        node_colors.append('blue')  # Open plants in blue
    else:
        node_colors.append('red')  # Customers in red

# Extract positions for graph nodes based on plant/customer positions
positions = {f'Customer {i}': customer_positions[i] for i in range(num_customers)}
positions.update({f'Plant {j}': plant_positions[j] for j in range(num_plants)})
```

```python
# Draw the bipartite graph
plt.figure(figsize=(10, 8))

# Draw nodes
nx.draw_networkx_nodes(B, positions, node_color=node_colors, node_size=700, alpha=0.9)

# Draw labels for nodes
labels = {node: node for node in B.nodes()}
nx.draw_networkx_labels(B, positions, labels, font_size=12)

# Display the plot
plt.axis("off")
plt.show()
```

```python
# Create a new model
model = gp.Model("plant_location")

# Decision variables
# y[j] = 1 if plant j is opened, 0 otherwise
y = model.addVars(num_plants, vtype=GRB.BINARY, name="y")

# x[i,j] = 1 if customer i is assigned to plant j, 0 otherwise
x = model.addVars(num_customers, num_plants, vtype=GRB.CONTINUOUS, name="x")

# Objective function: minimize fixed costs + transportation costs
model.setObjective(
    gp.quicksum(fixed_cost[j] * y[j] for j in range(num_plants)) +
    gp.quicksum(transport_cost[i][j] * x[i, j] for i in range(num_customers) for j in range(num_plants)),
    GRB.MINIMIZE
)

# Constraint 1: Each customer must be assigned to exactly one plant
model.addConstrs((gp.quicksum(x[i, j] for j in range(num_plants)) == 1 for i in range(num_customers)), name="customer_assignment")

# Constraint 2: Customers can only be assigned to open plants
model.addConstrs((x[i, j] <= 50 * y[j] for i in range(num_customers) for j in range(num_plants)), name="open_plant")
# Verbose
model.Params.LogToConsole = 0
# Optimize the model
model.optimize()

# Output solution details
if model.status == GRB.OPTIMAL:
    print("\nOptimal Solution Found:")
    print("\nPlant openings:")
    for j in range(num_plants):
        if y[j].x > 0.5:  # Plant j is opened
            print(f"Plant {j} is opened.")

    print("\nCustomer assignments:")
    for i in range(num_customers):
        for j in range(num_plants):
            if x[i, j].x > 0.5:  # Customer i is assigned to plant j
                print(f"Customer {i} is assigned to plant {j}.")
else:
    print("No optimal solution found.")
```

```python
# PLOT SOLUTION:
# Add edges for transportation cost, but only between open plants and assigned customers
edges = []
for i in range(num_customers):
    for j in range(num_plants):
        if x[i, j].x > 0.5:  # Customer i is assigned to plant j
            B.add_edge('Customer ' + str(i), 'Plant ' + str(j), weight=transport_cost[i][j])
            edges.append(('Customer ' + str(i), 'Plant ' + str(j)))

# Draw the bipartite graph
plt.figure(figsize=(10, 8))

# Draw nodes
nx.draw_networkx_nodes(B, positions, node_color=node_colors, node_size=700, alpha=0.9)

# Draw edges with weights (transportation costs)
nx.draw_networkx_edges(B, positions, edgelist=edges, width=2, edge_color="green", alpha=0.7)

# Draw labels for nodes
labels = {node: node for node in B.nodes()}
nx.draw_networkx_labels(B, positions, labels, font_size=12)

# Draw edge labels (transportation costs)
edge_labels = {(u, v): f'{B[u][v]["weight"]:.2f}' for u, v in edges}
nx.draw_networkx_edge_labels(B, positions, edge_labels=edge_labels, font_size=10)

# Display the plot
plt.title("Plant Location Model - Solution Representation with Distance-Based Costs")
plt.axis("off")
plt.show()
```

## Stochastic version
The stochastic model is then:

$$
\begin{array}
\text{min} \quad & \sum_{i \in \mathcal{P}} f_i y_i + \sum_{s\in\mathcal{S}}  \pi^s \! \left(\sum_{i \in \mathcal{P}} \sum_{j \in \mathcal{D}} c_{ij} x^s_{ij} \right) + \pi^s \! \left( \sum_{j \in \mathcal{D}} \hat{c}_{j} z_s^{j} \right) \\
\text{st} \quad & \sum_{i \in \mathcal{P}} x^s_{ij} \geq d^s_j - z_s^j \qquad \forall s \in \mathcal{S}, \; \forall j \in \mathcal{D} , \\
& \sum_{j \in \mathcal{D}} x^s_{ij} \leq R_i y_i \qquad \forall s \in \mathcal{S}, \; \forall i \in \mathcal{P} , \\
& x^s_{ij} \geq 0, \; y_i \in \{ 0,1 \} .
\end{array}
$$

```python
np.random.seed(1)
n_scenarios = 1
prob = 1 / n_scenarios
demand = np.random.normal(loc=10, scale=1, size=(num_customers,n_scenarios))
```

```python

model = gp.Model("plant_location")

# Decision variables
# y[j] = 1 if plant j is opened, 0 otherwise
y = model.addVars(num_plants, vtype=GRB.BINARY, name="y")

# x[i,j] = 1 if customer i is assigned to plant j, 0 otherwise
x = model.addVars(num_customers, num_plants, n_scenarios, vtype=GRB.CONTINUOUS, name="x")

# Objective function: minimize fixed costs + transportation costs
model.setObjective(
    gp.quicksum(fixed_cost[j] * y[j] for j in range(num_plants)) +
    prob * gp.quicksum(transport_cost[i][j] * x[i, j, s] for i in range(num_customers) for j in range(num_plants) for s in range(n_scenarios) ),
    GRB.MINIMIZE
)

# Constraint 1: Each customer must be assigned to exactly one plant
model.addConstrs(
    (gp.quicksum(x[i, j, s] for j in range(num_plants)) >= demand[i, s] for i in range(num_customers) for s in range(n_scenarios) ),
    name="customer_assignment"
)

# Constraint 2: Customers can only be assigned to open plants
model.addConstrs(
    (x[i, j, s] <= 20 * y[j] for i in range(num_customers) for j in range(num_plants) for s in range(n_scenarios)),
    name="open_plant"
)
# # Verbose
# model.Params.LogToConsole = 0
# Optimize the model
model.optimize()

if model.status == GRB.OPTIMAL:
    print("\nOptimal Solution Found:")
    print("\nPlant openings:")
    for j in range(num_plants):
        if y[j].x > 0.5:  # Plant j is opened
            print(f"Plant {j} is opened.")
    print("\nCustomer assignments:")
    for i in range(num_customers):
        for j in range(num_plants):
            for s in range(n_scenarios):
                if x[i, j, s].x > 0.5:  # Customer i is assigned to plant j
                    print(f"Customer {i} is assigned to plant {j} - x[{i}, {j}, {s}]")
model.write("model_stoch.lp")
```
